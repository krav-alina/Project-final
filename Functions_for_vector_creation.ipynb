```python
#предобрабатывает текст, делит на минимальные отрезки, сохраняет отрезки в отдельные файлы
for file in filenames:
    with open (file) as file1:
        tom = file1.read()
        punct = ",—»«][.…!?-:)("
        normalized_text = [morph.parse(word.strip(punct))[0].normal_form for word in tom.lower().split()]
        n = 1545
        split_text = [normalized_text[i:i+n] for i in range(0,len(normalized_text),n)]
        joined_text = [" ".join(i) for i in split_text]
        for number, item in enumerate(joined_text):
                with open(file +"_file_%s.txt" % number, "w") as f:
                    f.write(item)
            
```

```python
#открывает файл, записывает его в необходимый формат, считает вероятность для каждого сочетания
for file in filenames:
    header_list = ['words','numbers']
    table = pd.read_csv(file, engine = 'python', names = header_list)
    s = table.numbers.sum()
    
    for i in range(len(table.index)):
        probability = table.numbers[i]/s
        table['probability'] = pd.Series([probability for x in range(len(table.index))])
    table['text_name'] = pd.Series([file for x in range(len(table.index))])
                           
                    
    with pd.ExcelWriter('output4.xlsx', engine = "openpyxl", mode='a') as writer:  
        table.to_excel(writer)
```
